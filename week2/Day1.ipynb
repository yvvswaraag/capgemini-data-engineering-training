```sql
CREATE OR REPLACE TABLE Employees (
    emp_id INT PRIMARY KEY,
    emp_name VARCHAR(50),
    department_id INT,
    salary INT,
    manager_id INT
);

CREATE OR REPLACE TABLE Departments (
    dept_id INT PRIMARY KEY,
    dept_name VARCHAR(50)
);

CREATE OR REPLACE TABLE Orders (
    order_id INT PRIMARY KEY,
    emp_id INT,
    order_amount INT,
    order_date DATE
);
```

```sql
-- Find employees who earn more than average salary
select emp_name,salary 
from employees
where salary>(
  select avg(salary)
  from employees
)
```

```sql
-- Find employees who belong to the same department as 'Alice'
select *
from employees
where department_id=(
  select department_id
  from employees
  where emp_name='Alice'
)
```

```sql
-- Get employees whose salary is equal to the minimum salary
SELECT emp_name, salary 
FROM Employees 
WHERE salary = (
  SELECT MIN(salary) FROM Employees
  )
```

```sql
-- Find employees who have placed at least one order
SELECT emp_name
from employees
where emp_id in (
  SELECT emp_id
  from orders
)
```

```sql
-- Get employees whose salary is greater than ALL employees in department 102
select emp_name
from employees
where salary>(
  select max(salary)
  from employees
  where department_id=102
)
```

```sql
-- Create a CTE to show employees with salary > 50,000 and fetch all records
with emp as (
  select * from employees
  where salary>50000
)
select * from emp
```

```sql
-- Find average salary per department using CTE
with avg_sal as (
  select department_id, avg(salary) as avg_salary
  from employees
  group by department_id
)
select * from avg_sal
```

```sql
-- Use CTE to get employees and their department names
with emp_dept as (
  select e.emp_id, e.emp_name, e.salary, d.dept_name
  from employees e
  join departments d
    on e.department_id = d.dept_id
)
select * from emp_dept
```

```sql
-- Find total order amount per employee using CTE
with order_amount as (
  select emp_id, sum(order_amount) as total_order_amount
  from orders
  group by emp_id
)
select * from order_amount
```

```sql
-- Find employees whose salary is above department average using CTE
with deptavg as (
  select department_id,avg(salary) as avg_salary
  from employees
  group by department_id

)

select e.*
from employees e join deptavg d on e.department_id=d.department_id
where e.salary>d.avg_salary
```
