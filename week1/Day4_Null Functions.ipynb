```sql
select *
from employees
where salary is null
```

```sql
select *
from orders 
where discount is not null
```

```sql
select *
from products 
where category is null
```

```sql
select count(*) as emp_no_manager
from employees
where manager_id is null
```

```sql
select name,salary,ifnull(salary,0) as salary
from employees
```

```sql
select name,bonus,ifnull(bonus,1000)
from employees
```

```sql
select customer_name,amount,ifnull(amount,500) as amount
from orders
```

```sql
select stock,ifnull(stock,0)
from products
```

```sql
select name,salary,bonus,coalesce(salary,bonus,0) as earnings
from employees
```

```sql
select coalesce(salary,bonus,0) as earnings
from employees
```

```sql
select product_name,price,coalesce(price,1000)
from products
```

```sql
select customer_name,coalesce(amount,discount,0)
from orders
```

```sql
select name,salary,nullif(salary,0) as salary
from employees
```

```sql
select customer_name,discount,nullif(discount,0) as discount
from orders
```

```sql
select coupon_code,nullif(coupon_code,'DISC10')
from orders
```
