```python
from pyspark.sql import SparkSession
spark=SparkSession.builder.appName("Basics").getOrCreate()
df=spark.createDataFrame([("John",1),("Ram",2),("Siva",3)],
                         ["name","id"]
                         )
df.show()                         
```

```python
data = [
    ("Alice",   "HR",          50000),
    ("Bob",     "Engineering", 90000),
    ("Charlie", "HR",          55000),
    ("Diana",   "Engineering", 95000),
    ("Eve",     "Marketing",   60000),
    ("Frank",   "Marketing",   45000),
    ("Grace",   "Engineering", 85000)
]

columns = ["name", "department", "salary"]
df=spark.createDataFrame(data,columns)
df.show()
```


**SELECT() using column names and using col() object**

```python
df.select("name","salary").show()
```

```python
from pyspark.sql.functions import col
df.select(
    col("name").alias("emp_name"),
    col("department").alias("emp_dept")
).show()
```

```python
from pyspark.sql.functions import col
```

```python
df.select(
    col("name").alias("emp_name"),
    col("salary").alias("emp_salary"),
    (col("salary")/1000).alias("emp_salary_in_thousands")
).show()
```

**Filter() function**

```python
df.filter( col("salary")<60000).show()
```

```python
df.filter(col("department")=="Engineering").show()
```

```python
df.filter((col("department")=="Marketing")&(col("salary")>50000)).show()
```

```python
df.filter(~(col("department")=="Engineering")).show()
```

```python
df.filter(col("department").isin("HR","Marketing")).show()
```

**withColumn() add or modify the columns**

```python
df = df.withColumn("tax", col("salary") * 0.2)
```

```python

df=df.withColumn("net_salary",(col("salary")-col("tax")))
```

```python
df.show()
```

```python
from pyspark.sql.functions import lit
df=df.withColumn("company",lit("TechCrop"))
```

```python
df.show()
```

```python
from pyspark.sql.functions import when
df=df.withColumn(
    "experience_level",
    when(col("salary")>=90000,"Senior")\
    .when(col("salary")>=60000,"Mid")\
    .otherwise("Junior")
)
```

```python
df.show()
```

```python
from pyspark.sql.functions import upper
df.withColumn("name",upper(col("name"))).show()
```

```python
from pyspark.sql.functions import col, sum, avg, max, min, count
df.groupBy("department").agg(
    count("name").alias("employees_count")
).show()
```

```python
df.groupBy("department").agg(
    sum("salary").alias("Toatal_salary")
).show()
```

```python
df.groupBy("department").agg(
    max("salary").alias("Max_salary"),
    min("salary").alias("Min_salary")
).show()
```

```python
df.groupBy("department").agg(
    count("name").alias("total_employees"),
    sum("salary").alias("total_salary"),
    avg("salary").alias("average_salary"),
    max("salary").alias("max_salary")
).show()
```

```python
df.groupBy("department").agg(
avg("salary").alias("average_salary")).orderBy(col("average_salary").desc()).show()
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
spark = SparkSession.builder.appName("Joins").getOrCreate()
emp_data = [
    (1, "Alice",   101, 50000),
    (2, "Bob",     102, 90000),
    (3, "Charlie", 101, 55000),
    (4, "Diana",   102, 95000),
    (5, "Eve",     103, 60000),
    (6, "Frank",   104, 45000),
    (7, "Grace",   102, 85000)
]
emp_columns = ["emp_id", "name", "dept_id", "salary"]
emp_df = spark.createDataFrame(emp_data, emp_columns)
# DataFrame 2 — Departments
dept_data = [
    (101, "HR",          "Chennai"),
    (102, "Engineering", "Bangalore"),
    (103, "Marketing",   "Mumbai"),
    (105, "Finance",     "Delhi")
]
dept_columns = ["dept_id", "department", "location"]
dept_df = spark.createDataFrame(dept_data, dept_columns)
print("Employees:")
emp_df.show()
print("Departments:")
dept_df.show()
```

```python
joined_df=emp_df.join(dept_df,on="dept_id",how="inner")
joined_df.select("name","department","salary").show()
```

```python
joined_df=emp_df.join(dept_df,on="dept_id",how="left")
joined_df.filter(col("department").isNull()).show()
```

```python
joined_df=emp_df.join(dept_df,on="dept_id",how="right")
joined_df.filter(col("name").isNull()).show()
```

```python
joined_df=emp_df.join(dept_df,on="dept_id",how="full")
joined_df.count()
```

```python
joined_df=emp_df.join(dept_df,on="dept_id",how="left_anti")
joined_df.show()
```
