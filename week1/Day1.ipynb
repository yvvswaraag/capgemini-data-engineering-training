```python
df=spark.read.csv("/Volumes/workspace/default/databricks_training/C1.csv",header=True)
df.show()
```

```python
from pyspark.sql.functions import *

df=df.replace(["blank","Blank"],None)
df.display()
```

```python
df=df.withColumn("Country",upper(df["Country"]))
df.display()
```

##Handling missing values

##1.drop rows


```python
df=df.dropna(subset=["JoinDate"])
#dropna() removes all null values for clean processing
df.display()
```

##fill values

```python
df=df.fillna({
    "Category":"Unknown","Sales":0
})
df.display()
```

##3.droping duplicates

```python
df=df.dropDuplicates(["CustomerID"])
df.display()
```

##trim spaces

```python
from pyspark.sql.functions import trim
df.withColumn("Name",trim(col("Name")))
df.display();
```
