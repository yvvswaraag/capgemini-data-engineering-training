```sql
CREATE TABLE employee_payments (

emp_id INT PRIMARY KEY,

emp_name VARCHAR(50),

department VARCHAR(30),

base_salary DECIMAL(10,2),

bonus DECIMAL(10,2),

joining_date DATE

);

INSERT INTO employee_payments VALUES

(1,'karthik','Data',75000.75,5000.50,'2019-03-15'),

(2,'veena','HR',65000.40,4000.25,'2021-06-20'),

(3,'ravi','Data',85000.90,6000.75,'2016-01-10'),

(4,'anil','Finance',70000.10,NULL,'2020-09-01'),

(5,'suresh','HR',60000.55,3000.30,'2022-11-25');
```

```sql
select * from employee_payments;
```

```sql
select emp_id, round(base_salary+coalesce(bonus,0)) as total_income from employee_payments;
```

```sql
select emp_id, year(joining_date) as joining_year from employee_payments;
```

```sql
select emp_id,
    case 
        when year(curdate())-year(joining_date) > 7 then 'Senior'
        when year(curdate())-year(joining_date) between 4 and 7 then 'Mid'
        else 'junior'
        end as experience_level
from employee_payments
```

```sql
select user_id, user_email, substring_index(user_email, '@', -1) AS email_domain from subscriptions
```

```sql
select user_id, user_email, datediff(end_date, start_date)as duration from subscriptions
```

```sql
CREATE TABLE loan_details (

loan_id INT,

customer_name VARCHAR(50),

loan_amount DECIMAL(12,2),

interest_rate DECIMAL(5,2),

loan_start DATE

);


INSERT INTO loan_details VALUES

(201,'suresh',500000.75,8.5,'2022-01-10'),

(202,'mahesh',750000.40,9.2,'2021-05-20'),

(203,'anita',300000.90,7.8,'2023-07-01');
```

```sql
select * from loan_details;
```

```sql
select loan_id, upper(customer_name) as customer_name
from loan_details;
```

```sql
select loan_id, timestampdiff(year, loan_start, current_date) as years_passed
from loan_details;
```

```sql
select 
    loan_id,
    loan_amount * power(1 + interest_rate/100, 1) as monthly_interest
from loan_details;
```

```sql
select 
    loan_id,
    round(loan_amount * power(1 + interest_rate/100, 1)) as emi
from loan_details;
```

```sql
select 
    loan_id,
    case 
        when interest_rate > 9 then 'high risk'
        when interest_rate between 8 and 9 then 'medium risk'
        else 'low risk'
    end as risk_category
from loan_details;
```

```sql
CREATE TABLE attendance (

emp_id INT,

emp_name VARCHAR(50),

total_days INT,

present_days INT,

record_date DATE

);

INSERT INTO attendance VALUES

(1,'karthik',30,28,'2025-01-31'),

(2,'veena',30,22,'2025-01-31'),

(3,'ravi',30,18,'2025-01-31');

```

```sql
select emp_id, lower(emp_name) as emp_name
from attendance;
```

```sql
select 
    emp_id,
    round((present_days * 100.0) / total_days) as attendance_percentage
from attendance;
```

```sql
select 
    emp_id,
    monthname(record_date) as month_name
from attendance;
```

```sql
select 
    emp_id,
    (total_days - present_days) as absent_days
from attendance;
```

```sql
select 
    emp_id,
    case 
        when (present_days * 100.0) / total_days >= 90 then 'excellent'
        when (present_days * 100.0) / total_days between 75 and 89 then 'average'
        else 'poor'
    end as performance
from attendance;
```

```sql
CREATE TABLE product_sales (

product_id INT,

product_name VARCHAR(50),

mrp DECIMAL(10,2),

selling_price DECIMAL(10,2),

sale_date DATE

);


INSERT INTO product_sales VALUES

(1,'Laptop',75000.75,68000.50,'2025-01-10'),

(2,'Mobile',35000.40,33000.25,'2025-01-12'),

(3,'Tablet',25000.90,26000.75,'2025-01-15');
```

```sql
select * from product_sales;
```

```sql
select 
    product_id,
    concat(upper(substring(product_name,1,1)), lower(substring(product_name,2))) as product_name
from product_sales;
```

```sql
select 
    product_id,
    abs(mrp - selling_price) as discount_amount
from product_sales;
```

```sql
select 
    product_id,
    round(((mrp - selling_price) * 100.0) / mrp, 2) as discount_percentage
from product_sales;
```

```sql
select 
    product_id,
    dayname(sale_date) as day_name
from product_sales;
```

```sql
select 
    product_id,
    case 
        when selling_price < mrp then 'valid discount'
        when selling_price > mrp then 'overpriced'
        else 'no discount'
    end as discount_status
from product_sales;
```
