```python
clean_df = valid_sales \
    .join(customers, on="customer_id", how="left") \
    .join(cars, on="car_id", how="inner") \
    .join(sales_dealer, on="sale_id", how="left") \
    .join(dealers.withColumnRenamed("name", "dealer_name")
                 .withColumnRenamed("city", "dealer_city"), 
          on="dealer_id", how="left")

clean_df.display()
```

```python
clean_df = valid_sales.join(customers, on="customer_id", how="inner").join(cars, on="car_id", how="inner").join(sales_dealer, on="sale_id", how="left").join(dealers.withColumnRenamed("name", "dealer_name").withColumnRenamed("city", "dealer_city"),on="dealer_id", how="left")\
    .select(
        "customer_id",
        customers["name"].alias("customer_name"),  
        customers["city"].alias("customer_city"),  
        "sale_id",
        "car_id",
        "brand",
        "model",
        "price",
        "sale_date",
        "quantity",
        "dealer_name",
        "dealer_city",
        "dealer_id"
    )

clean_df.display()
```

**Transformations**

```python
clean_df=clean_df.withColumn("Customer revenue",(col("price")*col("quantity")))
clean_df.display()
```

```python
clean_df.groupBy("brand").agg(
    count(col("sale_id")).alias("Brand wise Sales")
).show()
```

```python
clean_df.groupBy("customer_city").agg(
    sum(col("Customer revenue")).alias("City wise Revenue")
).display()
```

```python
clean_df.groupBy("dealer_id").agg(
    sum(col("Customer revenue")).alias("Dealer wise Revenue")
).orderBy("dealer_id").display()
```

```python
total_revenue=clean_df.agg(sum(col("Customer revenue"))).collect()[0][0]
clean_df.groupBy("dealer_city").agg(
    round((sum(col("Customer revenue"))/total_revenue)*100,2).alias("Contribution")
).display()
```

```python
from pyspark.sql.window import Window
window=Window.partitionBy("customer_city").orderBy(col("Customer revenue").desc())
top_cus=clean_df.select("customer_name","customer_city",rank().over(window).alias("Top customers City Wise")).filter(col("Top customers City Wise")<=3).display()
```

```python
clean_df.groupBy(month(col("sale_date"))).agg(
    sum(col("Customer revenue")).alias("Total revenue"),
    count(col("sale_id")).alias("Total Sales"),
    avg(col("Customer revenue")).alias("Avg revenue")
).display()
```

```python
clean_df.groupBy("customer_id","customer_name").agg(
    count(col("sale_id")).alias("Total Purchases")
).filter(col("Total Purchases")>1).orderBy(col("Total Purchases").desc()).display()
```
